# Quick tutorial for using scotia module

```python
import scotia
import pandas as pd
import scanpy as sc
import anndata as ad
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import sys
```

```python
#set your workspace directory
datap = '/data/hemberg/jingyi/project/1.pdac/scripts/dbscan_ot_py/'
```

### load input

```python
adata_test = sc.read_h5ad(datap+'example/merfish_liver_example.h5ad')
adata_test
```

### load known ligand recepter pair

```python
#known LR pairs
known_lr_pairs = pd.read_csv(datap+'example/lr_gene.list', sep = '\t', header = None, index_col = None)
known_lr_pairs.columns = ['l_gene','r_gene']
print(known_lr_pairs.head())
```

### Run scotia

```python
scotia.run_scotia.lr_score(adata = adata_test,
         lr_list = known_lr_pairs,
         sample_col = 'sample',
         fov_col = 'fov',
         celltype_col = 'annotation',
         output_path = datap+'example/test_output/')
```

### Check output
